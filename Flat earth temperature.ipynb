$I(r,\theta) = \frac{S(r_s, \theta_s, z_s)}{4\pi(x-x_s)^2}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
R = 1
zs = 0.5
rs = 0.5
```

```python
fig = plt.figure()
ax = Axes3D(fig)

rs = 0.5
zs = 0.5
ths = 0

I_sun = 0
rad = np.linspace(0, 1, 1000)
azm = np.linspace(0, 2 * np.pi, 1000)
r, th = np.meshgrid(rad, azm)
for i in np.linspace(0,2 * np.pi, 100):
    I_sun += 1/(4*np.pi*(np.square(zs) + np.square(r) + np.square(rs) - 2*r*rs*np.cos(th-i)))

plt.subplot(projection="polar")

plt.pcolormesh(th, r, I_sun, cmap="jet")
#plt.pcolormesh(th, z, r)

plt.plot(azm, r, color='k', ls='none') 
plt.grid()

plt.title("Average heat from sun")
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)

rs = 0.5
zs = 0.5
ths = 0

I_sun_season = 0
rad = np.linspace(0, 1, 1000)
azm = np.linspace(0, 2 * np.pi, 1000)
r, th = np.meshgrid(rad, azm)
for j in np.linspace(-0.1,0.1,10):
    for i in np.linspace(0,2 * np.pi, 100):
        I_sun_season += 1/(4*np.pi*(np.square(zs) + np.square(r) + np.square(rs+j) - 2*r*(rs+j)*np.cos(th-i)))

plt.subplot(projection="polar")

plt.pcolormesh(th, r, I_sun_season, cmap="jet")
#plt.pcolormesh(th, z, r)

plt.plot(azm, r, color='k', ls='none') 
plt.grid()
plt.title("Average from sun heat with seasons")
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)

rs = 0.5
zs = 0.5
ths = 0

I = 0
rad = np.linspace(0, 1, 1000)
azm = np.linspace(0, 2 * np.pi, 1000)
r, th = np.meshgrid(rad, azm)
for i in np.linspace(0,2 * np.pi, 100):
    d_projected = np.sqrt(np.square(r) + np.square(rs) - 2*r*rs*np.cos(th-i))
    angle = np.arctan(d_projected/zs)
    I += 1/(4*np.pi*(np.square(zs) + np.square(r) + np.square(rs) - 2*r*rs*np.cos(th-i)))*(angle<np.pi/4)

plt.subplot(projection="polar")

plt.pcolormesh(th, r, I_sun_season, cmap="jet")
plt.title("Average heat from spotlight")
plt.plot(azm, r, color='k', ls='none') 
plt.grid()

plt.show()
```
